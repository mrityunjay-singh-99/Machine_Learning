```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
l=[]
for i in range(len(input_df.columns)):
    l.append(variance_inflation_factor(input_df.values,i))
VIF_data=pd.DataFrame()
VIF_data["features"]=input_df.columns
VIF_data["VIF"]=l
VIF_data
```

```python
# Case1>>>>first we perfrom Scaler on datasets then VIF.

vif_threshold=15
features_to_keep=VIF_data[VIF_data["VIF"]<vif_threshold]["features"].to_list()

califonia_housing_new_df=califonia_housing_df[features_to_keep+["MedHouseVal"]]
califonia_housing_new_df
```
